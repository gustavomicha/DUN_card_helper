```python
crafting_hardcoded = crafting.copy()

crafting_hardcoded = crafting_hardcoded.applymap(lambda x: x.replace("Alquimista", "assets\\misc\\alquimista.png") if isinstance(x, str) else x)
crafting_hardcoded = crafting_hardcoded.applymap(lambda x: x.replace("Escuela_de_hechiceria", "assets\\misc\\esc_hechiceria.png") if isinstance(x, str) else x)
crafting_hardcoded = crafting_hardcoded["Resultado"] = crafting_hardcoded["Resultado"].apply(
    lambda x: f"assets\images\ESP\items\{x}.png" if isinstance(x, str) and not x.startswith("objeto_") else x
)
crafting_hardcoded = crafting_hardcoded["recurso original	"] = crafting_hardcoded["recurso original	"].apply(
    lambda x: f"assets\misc\{x}.png" if isinstance(x, str) and not x.startswith("objeto_") else x
)
```

```python
def crafting_combinations(df, recurso_original):
    df_filt = df[df["recurso original"] == recurso_original]
    list_of_dict = []
    for _, row in df_filt.iterrows():
        r1 = row['recurso original']
        r2 = row['recurso']
        extra = row['Extra']
        oro = row['Oro']
        lugar = row['Lugar']
        item = row['Resultado']
        item_dict = {
            'recurso_original':r1, 
            'recurso':r2,
            'Extra':extra,
            'Oro':oro,
            'Lugar':lugar,
            'Resultado':item,
        }
        print(f"Recurso {r1} + {r2} {'+ ' + extra if pd.notna(extra) else ''}+ {oro} oro en {lugar} genera {item}")
        list_of_dict.append(item_dict)

    return list_of_dict
```
